```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL base de la página
base_url = 'https://www.walmart.com.gt/abarrotes?page={}'

# Número máximo de páginas que deseas analizar
numero_de_paginas = 10  # Puedes ajustar esto según tus necesidades

# Clase específica del cuadro que deseas extraer
clase_cuadro = 'vtex-product-summary-2-x-container vtex-product-summary-2-x-containerNormal overflow-hidden br3 h-100 w-100 flex flex-column justify-between center tc'

# Crear listas para almacenar la información
nombres_productos = []
precios_productos = []
marcas_productos = []

# Bucle a través de las páginas
for pagina in range(1, numero_de_paginas + 1):
    # Construir la URL para la página actual
    url = base_url.format(pagina)

    # Realizar la solicitud GET a la página web
    response = requests.get(url)

    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Parsear el contenido de la página con BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrar elementos con la clase especificada
        elementos_con_clase = soup.find_all(class_=clase_cuadro)

        # Extraer la información de cada elemento
        for elemento in elementos_con_clase:
            nombre_producto = elemento.find('a', class_='product-name')
            precio_producto = elemento.find(class_='vtex-store-components-3-x-sellingPrice vtex-store-components-3-x-sellingPriceContainer pv1 b c-on-base vtex-store-components-3-x-price_sellingPriceContainer vtex-store-components-3-x-price_sellingPriceContainer--summary')
            marca_producto = elemento.find(class_='vtex-product-summary-2-x-productBrand vtex-product-summary-2-x-brandName t-body')

            # Verificar si los elementos existen antes de acceder a sus atributos


            if precio_producto is not None:
                precio_producto = precio_producto.text.strip()
            else:
                precio_producto = "No disponible"

            if marca_producto is not None:
                marca_producto = marca_producto.text.strip()
            else:
                marca_producto = "No disponible"

            # Agregar la información a las listas respectivas
            nombres_productos.append(nombre_producto)
            precios_productos.append(precio_producto)
            marcas_productos.append(marca_producto)
    else:
        print(f'Error al acceder a la página {pagina}. Código de estado: {response.status_code}')

# Crear un DataFrame con la información recopilada
data = {
##    'Nombre del Producto': nombres_productos,
    'Precio del Producto': precios_productos,
    'Marca del Producto': marcas_productos
}

df = pd.DataFrame(data)

# Imprimir el DataFrame
print(df)
```
